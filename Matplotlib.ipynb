# Matplotlib Library

```python
# import matplotlib module
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()

plt.plot([1, 2, 3], [1, 2, 3])
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

line = axes[0,0].plot([1,2,3],[1,2,3])
axes[0,1].scatter([1,2,3],[1,2,3])
axes[1,0].bar([1,2,3],[1,2,3])
axes[1,1].hist([1,1,1,2,2,3,3,3,3], bins=3)

plt.show()
```

```python

plt.plot([1,2,3,4,5], [10,20,30,35,25])


plt.xlim(0, 6)
plt.ylim(0, 40)
plt.xlabel('X-label')
plt.ylabel('Ylabel')
plt.title('title')
plt.show()
```

```python
fig, ax = plt.subplots()

line = ax.plot([1,2,3],[4,5,6], label='line')[0]
#modidy the line
line.set_color('green')
line.set_linewidth(3.5)
# add labels and title
ax.set_xlabel('X-label')
ax.set_ylabel('Y-label')
ax.set_title('Artist label')
plt.legend()
plt.show()
```
