# Matplotlib Library

```python
# import matplotlib module
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()

plt.plot([10,20,30,40,50],[0.5,1.0,1.5,2.0,2.5])
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

line = axes[0,0].plot([30,20,30,40,50],[0.5,1.5,1.5,2.0,1.5])
axes[0,1].scatter([10,20,30,40,50],[0.5,1.5,1.5,2.0,1.5])
axes[1,0].bar([10,20,30,40,50],[0.5,1.0,1.5,2.0,2.5])
axes[1,1].hist([10,20,30,40,50,10,20,30,40,50])
plt.show()
```

```python

plt.plot([10,20,30,40,50],[10,20,30,40,50], label="graph", color="red")

plt.xlim(0,100)
plt.ylim(0,100)
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.title("Graph")
plt.legend()
plt.show()
```
